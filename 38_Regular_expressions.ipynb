# Regex (Regular Expressions)

```python
import re
```

## re.search()

```python
text = """Hello World.
I love Python!!!!!
My first lucky number is 987-654-321
My second lucky number is 876-543-210
My third lucky number is 765-432-100
My fourth lucky number is 123-456-123-123-456
"""
```

```python
# return the first match object
re.search(r'\d', text)
```

```python
# Flags
re.search(r'^hello', text, re.IGNORECASE)  # case insensitive search
```

## re.findall()

```python
# return all the matches as a list of strings
re.findall(r'\d', text)
```

## Exercises

```python
# Match only punctuation
text = """Hello World.
I love Python!!!!!
My first lucky number is 987-654-321
My second lucky number is 876-543-210
My third lucky number is 765-432-100
My fourth lucky number is 123-456-123-123-456
"""

re.findall(r'[^\w\s]', text)    # This will match only the punctuation marks
```

```python
# match right data format (MMM-DDD-YYYY)
text = '''
13-04-2021
2021-13-04
2021-04-13
'''
re.findall(r'\d{2}-\d{2}-\d{4}', text) 
# This will match only the dates in DD-MM-YYYY format 
```

```python
# Match right username format (4-14 characters and only letter or digits)
text = '''
username10
abc
10
username_1234567890
'''
re.findall(r'\b[a-zA-Z0-9]{4,14}\b', text)
```
