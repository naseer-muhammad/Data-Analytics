```python
import pandas as pd
```

```python
# Reads csv file with proper encoding
df = pd.read_csv("players_20.csv")
# set index 
df.set_index("short_name", inplace = True)
# Select Column
df = df[["long_name", "age", "dob", "height_cm", "weight_kg", "nationality"]]
```

```python
# show dataframe
df
```

# Drop Row

```python
# Drop one row
# Axis parameter
df.drop("L. Messi", axis = 0) 
# axis = 0: Row, axis = 1: Column

# Index parameter
df.drop(index = ["L. Messi"])
```

```python
# Drop two or more rows and update data (inplace = True)
df.drop(["L. Messi", "Cristiano Ronaldo"], axis = 0, inplace = True)
```

```python
df
```

# Drop Columns

```python
# Drop one column
# Axis parameter
df.drop("long_name", axis = 1) 
# axis = 0: Row, axis = 1: Column

# Index parameter
df.drop(columns = ["long_name"])
```

```python
# drop colum by position/index (drop last column)
df.drop(df.columns[[1]], axis = 1)
```

```python
# dorp two or more columns and update data (inplace = True)
df.drop(["long_name", "dob"], axis = 1, inplace = True) 
```
