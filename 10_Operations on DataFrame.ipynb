```python
import pandas as pd
```

```python
# Reading the CSV file 
df_exams = pd.read_csv("StudentsPerformance.csv")
df_exams.head()
```

# Math Operations

## Operations in Columns

```python
# Select a column that you want to add and calculate total sum 
df_exams["math score"].sum()
```

```python
# Count, Mean, std (Standard Deviation), Max, Min
print(df_exams["math score"].count())
print(df_exams["math score"].mean())
print(df_exams["math score"].sum())
print(df_exams["math score"].std())
print(df_exams["math score"].max())
print(df_exams["math score"].min())
```

```python
# Easier way to calculate with .describe()
df_exams.describe().round(2)
```

## Operations in Rows

```python
# Calculating the sum in a row
df_exams["math score"] + df_exams["reading score"]	+ df_exams["writing score"]
```

```python
# Calculating the average score and assigning the result to a new columnd
df_exams["average"] = (df_exams["math score"] + df_exams["reading score"] + df_exams["writing score"])/3
```

```python
# Showing the dataframe
df_exams.round(2)
```
