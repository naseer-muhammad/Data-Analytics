```python
import pandas as pd
```

```python
# Reads csv file with proper encoding
df = pd.read_csv("players_20.csv")
# set index 
df.set_index("short_name", inplace = True)
# Select Column
df = df[["long_name", "age", "dob", "height_cm", "weight_kg", "nationality"]]
```

```python
# show dataframe
df
```

# sample()

```python
# extract 10 random elements from the "nationality" column
df["nationality"].sample(11, random_state = 99)
```

```python
# Extract a random 20% sample of the dataframe
df.sample(frac=0.2, random_state = 12)
```

```python
# Unsample: increase the sampling rate -> frac > 1 (Note: replace parameter has to be True for frac parameter > 1)
df.sample(frac=2, replace = True, random_state= 0)
```
