# Create a column based on 1 conditions: np.where()

```python
import numpy as np
```

```python
# Create an array based on price tier
np.where(df_laptops["Price_euros"] >  2000, "Expensive", "Cheap")
                                            # For True,  for False
```

```python
# Set the above to a new columm
df_laptops["Price_tier"] = np.where(df_laptops["Price_euros"] >  2000, "Expensive", "Cheap")
```

```python
# show first 5 rows
df_laptops.head()
```

```python
# COunt the values in "price_tier" column
df_laptops.value_counts("Price_tier")
```
