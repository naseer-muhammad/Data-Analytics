```python
import pandas as pd
```

```python
# Reading the CSV file 
df_exams = pd.read_csv("StudentsPerformance.csv")
df_exams
```

# Create Index

```python
import numpy as np
import random
```

```python
# Creating non repeptitive values for the index
new_index = np.arange(0, 1000)
```

```python
# shuffling the indexes created above
random.shuffle(new_index)
```

```python
# Creating a new colum with new index
df_exams["new_index"] = new_index
df_exams
```

## Set index

```python
df_exams.set_index("new_index", inplace=True)
df_exams
```

## Sort index

```python
# Sort dataframe by new index
df_exams.sort_index()
```

```python
# Sort dataframe by new index by decenging
df_exams.sort_index(ascending = False)
```

```python
# To save the changes, we can use the inplace
df_exams.sort_index(ascending = False, inplace=True)
```
