```python
import pandas as pd
```

```python
df_laptops = pd.read_csv("laptop_price.csv", encoding = "latin1")
```

```python
df_laptops.head()
```

# isin(): Single Filtering

```python
# Select Apple or HP Laptops
df_laptops["Company"].isin(["Apple", "HP"])
```

```python
# Filter Dataframe
df_laptops[df_laptops["Company"].isin(["Apple", "HP"])]
df_laptops[df_laptops["Company"].isin(["Apple", "HP"])].value_counts("Company")
```

# isin(): Multiple Filterning

```python
# Find notebook or ultrabook from Apple or HP
filter1 = df_laptops["TypeName"].isin(["Notebook", "Ultrabook"])
```

```python
filter2 = df_laptops["Company"].isin(["Apple", "HP"])
```

```python
# Filter dataframe
df_laptops[filter1 & filter2]
```
