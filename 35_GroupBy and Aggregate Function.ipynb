# 2. The split-Apply-Combine Strategy

```python
df_cars = pd.read_csv("Car_sales.csv")
df_cars
```

```python
# split data into seperate groups
car_filter = df_cars['Vehicle_type']=='Car'
passenger_filter = df_cars['Vehicle_type']=='Passenger'
# df_cars['Vehicle_type'].value_counts()
```

```python
# Apply an operation
car_average = df_cars[car_filter]['Sales_in_thousands'].mean()
passenger_average = df_cars[passenger_filter]['Sales_in_thousands'].mean()
```

```python
# combine results
pd.DataFrame({'Vehicle_type':['Car', 'Passenger'],
              'Sales_in_thousands':[car_average, passenger_average]})

```

```python
# We can also make teh Vehicle_type to be the index
pd.DataFrame({'Vehicle_type':['Car', 'Passenger'],
              'Sales_in_thousands':[car_average, passenger_average]}).set_index("Vehicle_type")
```

# 4. Groupby() and agg()

```python
# find the minimum and maximum values on each column

# df_cars.groupby('Vehicle_type').max()
# df_cars.groupby('Vehicle_type').mmin()

# or we can do the same with the one line. We can do that to as many agg("function") as you want to.
# We can also customise different names of the columns

df_cars.groupby('Vehicle_type').agg(['min', 'max'])
```

```python
# set a name for aggregated column
# find the min Engine_size and max Horsepower
df_cars.groupby('Vehicle_type').agg(min_engine_size=('Engine_size', 'min'),
                                    max_horsepower=('Horsepower', 'max'))
```

Apply your own function

```python
df_cars = df_cars[["Sales_in_thousands","Price_in_thousands",
                                    "Engine_size","Horsepower", "Fuel_capacity"]]
df_cars
```

```python
# First reload the original dataframe
df_cars = pd.read_csv("Car_sales.csv")

# Then run your groupby operation
df_cars.groupby('Manufacturer').sum()[['Sales_in_thousands', 'Price_in_thousands']].apply(lambda x:x*1000).head()
```

```python
# Select only numeric columns for the calculation 
numeric_cols = ['Sales_in_thousands', 'Price_in_thousands', 'Engine_size', 'Horsepower', 'Fuel_capacity']

# Subtract mean from each group using only numeric columns
result = df_cars.groupby('Manufacturer')[numeric_cols].apply(lambda x: x - x.mean())
result
```
