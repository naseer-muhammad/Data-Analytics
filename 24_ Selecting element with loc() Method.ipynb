```python
import pandas as pd
```

```python
# read csv file
df = pd.read_csv("players_20.csv")
# set index
df.set_index("short_name", inplace=True)
# Select columns
df = df[["long_name", "age", "dob", "height_cm", "weight_kg", "nationality", "club"]]
df
```

```python
# Get all data about the L.Messi
df.loc["L. Messi"]
```

```python
# Get height of L.Messi
df.loc["L. Messi", "height_cm"]
```

```python
# get weight of Cristiano Ronaldo
df.loc["Cristiano Ronaldo", "weight_kg"]
```

```python
# get all the rowws inside the "height_cm" column
df.loc[:, "height_cm"]
```

```python
# get all coulmns that correspond to the index "L. Messi"
df.loc["L. Messi", :]
```

# Selecting with a list of values

```python
# Get all data about L. Messi and Cristiano Ronaldo

df.loc[["L. Messi", "Cristiano Ronaldo"]]
```

```python
# Get the height of L. Messi and Cristiano Ronaldo
df.loc[["L. Messi", "Cristiano Ronaldo"] , "height_cm"]
```

```python
# get the height and weigh of L. Messi
df.loc["L. Messi", ["weight_kg", "height_cm"]]
```

```python
# Get height and weight of L. Messi and Cristiano Ronaldo 
df.loc[["L. Messi", "Cristiano Ronaldo"], ["weight_kg", "height_cm"]]
```

# Select a range of data with a slice
start:stop:step (note that contrary to usual python slices, both the start and the stop are included)

```python
# Slice column labels
players = ["L. Messi", "Cristiano Ronaldo"]

df.loc[players, "age": "club"]
```

```python
# Slice index labels
columns = ["age", "dob", "height_cm", "weight_kg"]

# Get top1 and top10 player name
df.index[:10]

df.loc['L. Messi':'M. Salah', columns]
```

# Selecting with conditions

```python
# one codition: Select player with height above 180cm
columns = ["age", "dob", "height_cm", "weight_kg"]

df.loc[df["height_cm"]> 180, columns]
```

```python
# Multiple condiitonsL: select players with height above 180 from Argentina
df.loc[(df["height_cm"]> 180) & df.loc[df["nationality"]== "Argentina", :]
```
