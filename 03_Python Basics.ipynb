```python
# To get the keys of the dictionary
my_data.keys()
```

```python
my_data.values()
```

```python
# An item in dictionary is combination of 'key' and 'value'. For Example
my_data.items()
```

```python
# TO add more detials into the dictionary
my_data["height"] = 1.7
```

```python
my_data
```

```python
# We can update dictionary also using .update() method
my_data.update({"height": 1.8})
my_data
```

# IF Statments

```python
'''Syntax of IF Statements'''
# if <condition>:
#     <code>

# elif <condition>:
#     <code>
# ...
# else:
# <code>
```

```python
age = 11

if age >= 18:
    print("You are of legal age adult.")

elif age>= 13:
    print("You are a teenager.")
else: 
    print("You are a kid.")
```
