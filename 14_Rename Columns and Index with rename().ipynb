```python
import pandas as pd
```

```python
# Reading the CSV file 
df_exams = pd.read_csv("StudentsPerformance.csv")
df_exams.head()
```

# Rename the Column

```python
# Rename column and overwrite values of the dataframe
df_exams = df_exams.rename(columns={"gender": "Gender"})
```

# Rename Index

```python
# Rename index 0, 1 and 2 and update the dataframe
df_exams.rename(columns={"math score": "MS", "reading score": "RS", "writing score": "WS"}, inplace= True)
df_exams
```

# Rename Index (by Index argument)

```python
# Rename index 0, 1 and 2 and update the dataframe
df_exams.rename(index={0: "A", 1: "B", 2: "C"}, inplace= True)
df_exams
```
