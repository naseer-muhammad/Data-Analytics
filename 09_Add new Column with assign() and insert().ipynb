# 1. Adding a New Column
### 1.1. assign()

When to use assign?
* Add multiple columns in a single line of code
* When you need to overwrite the values of an exisiting column (best practice).
  It returns a new object (a copy) with all the original columns in addition to the new ones.

```python
import numpy as np
```

```python

# Create random numbers for our new score 'columns' 
Physics = np.random.randint(33, 100, size= 1000)
Chemisty = np.random.randint(33, 100, size= 1000)
```

```python
series1 = pd.Series(Physics, index= np.arange(0, 1000))
series2 = pd.Series(Chemisty, index= np.arange(0, 1000))
```

```python
# Using assign() to add multiple columns 
#assign(Name of the column= values, Name of the column= values)
df_exams = df_exams.assign(Physics=series1, Chemisty=series2)
```

```python
# Show dataframe
df_exams
```

### 1.2. insert()

```python
# using index() to add a column at a specific position
# inset (the index where you want the new column to be, name of the column, the column whose values you want to copy)
df_exams.insert(5, "test_new", series1)
```
